```python
from pytrends.request import TrendReq
import pandas as pd
import time
startTime = time.time()
pytrend = TrendReq(hl='en-GB', tz=360)

colnames = ["keywords"]
df = pd.read_csv("C:\\Users\\paulo.almeida\\Documents\\CSV\\keyword_list.csv", names=colnames)
df2 = df["keywords"].values.tolist()
df2.remove("Keywords")

#dataset = []
print('Carregando...aguarde')
dataset =[]
anos = [2005,2006,2007,2008]#,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
meses = ['01','02','03','04','05','06','07','08','09','10','11','12']


pytrends = TrendReq(hl='en-US', tz=360, timeout=(10,25), retries=2, backoff_factor=0.1)#proxies=['https://34.203.233.13:80',]
for ano in anos:
    
   
    
    print(ano)
    for mes in meses:
        print(mes)
        if mes == '02':
            dia_final=28
        else: dia_final = 30
        
        if mes == '03':
            print('Sleep_mar√ßo')
            time.sleep(1200)
        elif mes == '06':
            print('Sleep_junho')
            time.sleep(1200)
        elif mes == '10':
            print('Sleep_outubro')
            time.sleep(1200)
        else: pass
        for x in range(0,len(df2)):
             
             
             timeframes='{}-{}-01 {}-{}-{}'.format(ano,mes,ano,mes,dia_final)
             keywords = [df2[x]]
             pytrend.build_payload(
             kw_list=keywords,
             cat=0,
             timeframe=timeframes,
             geo='GB')
             print(timeframes)
             time.sleep(3)
             executionTime = (time.time() - startTime)
             #print('Execution time in sec.: ' + str(executionTime))
             data = pytrend.interest_over_time()

             if not data.empty:
                  data = data.drop(labels=['isPartial'],axis='columns')
                  dataset.append(data) 
     
             print(keywords)
             
             
        time.sleep(15)
        
          
    result = pd.concat(dataset, axis=1)
    result.to_csv("C:\\Users\\paulo.almeida\\Documents\\CSV\\Resultado_GTrends{}.csv".format(ano))
    executionTime = (time.time() - startTime)
    print('Execution time in sec.: ' + str(executionTime))
    print('Sleep_dezembro')
    time.sleep(2000)

result = pd.concat(dataset, axis=1)
executionTime = (time.time() - startTime)
```
